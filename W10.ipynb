```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
from sklearn import datasets,metrics
```

```python
x_axis = np.array([1,3,5,9,12,16,23])
y_axis = np.array([2,4,6,8,10,12,16])

data = pd.DataFrame({'x':x_axis, 'y':y_axis})

plt.plot()
plt.xlim([0,21])
plt.ylim([0,21])
plt.title('Dataset')
plt.scatter(x_axis,y_axis)
plt.show()
```

```python

print('For ONE Cluster, all data points should be assigned to 1 Cluster\n')

kmeans = KMeans(n_clusters = 1, max_iter = 100).fit(data)
data['clusters'] = kmeans.labels_
print('Clusters assigned to each datapoints :')
print(data['clusters'])
print('\nSum of Square Error/ WCSS is :',kmeans.inertia_)
```

```python

print('For ONE Cluster, all data points should be assigned to 2 Cluster\n')

kmeans = KMeans(n_clusters = 2, max_iter = 100).fit(data)
data['clusters'] = kmeans.labels_
print('Clusters assigned to each datapoints :')
print(data['clusters'])
print('\nSum of Square Error/ WCSS is :',kmeans.inertia_)
```

```python


# ELBOW METHOD

# How many clusters should we choose ? - Elbow Method
# Select the elbow point for number of clusters
data = pd.DataFrame({'x':x_axis, 'y':y_axis})

sse = {}

for k in range(1,6): # Maximum range should be 6, as it contains only 6 data points
  kmeans = KMeans(n_clusters=k,max_iter=100).fit(data)
  data['clusters'] = kmeans.labels_
  sse[k] = kmeans.inertia_
  print("For cluster = {}, SSE/WCSS is {}".format(k, sse[k]))

plt.figure()
plt.plot(list(sse.keys()),list(sse.values()))
plt.xlabel("Number of Cluster")
plt.ylabel("SSE / WCSS")
plt.title("ELBOW METHOD")
plt.show()

print('Optimum Number of Cluster :2 ')
     
```

```python

# Silhouette Score

data = pd.DataFrame({'x':x_axis, 'y':y_axis})

for n_clusters in range(0,6):# Maximum range should be 6, as it contains only 6 data points
  kmeans = KMeans(n_clusters=k,max_iter=100).fit(data)
  label = kmeans.labels_
  sil_coeff = silhouette_score(data,label,metric = 'euclidean')
  print('For cluster= {}, Silhouette Coefficient is {}'.format(n_clusters,sil_coeff))

print('\n It is same for every cluster, so choose from Elbow Method')
```

## Week 10 on Iris Dataset

# 1. Import libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
```

```python
# 2. Load the dataset
```

```python
iris = load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)
```

```python
# 3. Look at top 5 values
```

```python
print("Top 5 rows of the dataset:\n", df.head())
```

```python
# 4. Select only 2 features (e.g., sepal length and petal width for simplicity)
```

```python
data = df.iloc[:, [0, 3]] 
print(data)
```

```python
# 5. Import KMeans 
from sklearn.cluster import KMeans
```

```python
# 6. List to store within-cluster sum of squares (WCSS)
```

```python
wcss = []
```

```python
# 7. Fit KMeans with different values of K (1 to 10)
```

```python
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', max_iter=300, random_state=42)
    kmeans.fit(data)
    wcss.append(kmeans.inertia_)
    print(f"WCSS for k={i}: {kmeans.inertia_}")
```

```python
# 8. Plot the WCSS values
```

```python
plt.plot(range(1, 11), wcss, marker='o')
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.grid()
plt.show()
```

```python
# 9. Automatically detect optimal number of clusters
```

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=5, init='k-means++', max_iter=300, random_state=42)
y_kmeans = kmeans.fit_predict(data)
```

```python
# 10. Fit KMeans again with optimal K
```

```python
print("Cluster labels assigned to each data point:")
print(y_kmeans)
```

```python
# 11. Visualizing the clusters
```

```python
import matplotlib.pyplot as plt

# Assign different colors for each cluster
colors = ['red', 'blue', 'green', 'cyan', 'magenta']
for i in range(5):
    plt.scatter(data.iloc[y_kmeans == i, 0], data.iloc[y_kmeans == i, 1],
                s=100, c=colors[i], label=f'Cluster {i+1}')

# Plot the centroids
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1],
            s=300, c='yellow', marker='X', label='Centroids')

plt.title('Clusters of Iris Dataset (K=5)')
plt.xlabel('Feature 1')  # Replace with actual column name like 'SepalLengthCm'
plt.ylabel('Feature 2')  # Replace with actual column name like 'PetalWidthCm'
plt.legend()
plt.grid(True)
plt.show()
```
